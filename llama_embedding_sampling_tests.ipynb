```python
from llama_models import full_vocab_embedding
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Distribution of per-dimension mins and maxes
mins_and_maxes = np.stack([np.min(full_vocab_embedding, axis = 0),
np.max(full_vocab_embedding, axis = 0),
np.mean(full_vocab_embedding, axis = 0),
np.std(full_vocab_embedding, axis = 0)])
sns.histplot(mins_and_maxes[0,:], label = "Dimension Minimums")
sns.histplot(mins_and_maxes[1,:], label = "Dimension Maximums")
sns.histplot(mins_and_maxes[2,:], label = "Dimension Means")
sns.histplot(mins_and_maxes[3,:], label = "Dimension StDev")
plt.legend()
plt.show()
```
